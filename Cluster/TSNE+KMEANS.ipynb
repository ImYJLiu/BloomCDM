```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 加载数据集
def load_data(data_path):
    points = np.loadtxt(data_path)
    return points
```

```python
data = load_data('../know_user_kmeans.txt')
print(data.shape)
```

```python
# 使用TSNE进行降维处理。从1000维降至2维。
tsne = TSNE(n_components=2, learning_rate=50).fit_transform(data)
```

```python
print(tsne.shape)
```

```python
plt.figure(figsize=(10, 10))
plt.plot(2)
plt.scatter(tsne[:, 0], tsne[:, 1])
plt.show()
```

```python
from sklearn.cluster import Birch
y_pred = Birch(n_clusters = 2,threshold = 0.3, branching_factor = 20).fit_predict(tsne)
plt.scatter(tsne[:, 0], tsne[:, 1], c=y_pred)
plt.title('Estimated number of clusters: %d' % len(set(y_pred)))
plt.show()
```
